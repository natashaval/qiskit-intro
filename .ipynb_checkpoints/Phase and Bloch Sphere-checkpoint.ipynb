```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram

%matplotlib inline
import math
```

```python
qasm_simulator = Aer.get_backend('qasm_simulator')
state_simulator = Aer.get_backend('statevector_simulator')
```

```python
def run_on_simulator(circuit):
    statevec_job = execute(circuit, backend=state_simulator)
    result = statevec_job.result()
    statevec = result.get_statevector()
    
    num_qubits = circuit.num_qubits
    circuit.measure([i for i in range(num_qubits)], [i for i in range(num_qubits)])
    
    qasm_job = execute(circuit, backend = qasm_simulator, shots = 1024).result()
    qasm_counts = qasm_job.get_counts()
    
    return statevec, qasm_counts
```

```python
circuit = QuantumCircuit(2,2)
statevec, counts = run_on_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
circuit.h(0)
circuit.draw(output='mpl')
```

```python
statevec, counts = run_on_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
plot_histogram(counts)
```

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0,1)
statevec, counts = run_on_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
circuit = QuantumCircuit(2,2)
circuit.ry(math.pi/4, 0)
circuit.ry(math.pi/2, 1)
statevec, counts = run_on_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
circuit = QuantumCircuit(1,1)
circuit.h(0)
circuit.z(0)
statevec, counts = run_on_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
plot_histogram(counts)
```
