```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
%matplotlib inline

circuit = QuantumCircuit(3,3)
# change value of Q0 to '1'
# circuit.x(0)
circuit.h(0)
circuit.barrier()

# create entanglement Q1 and Q2
circuit.h(1)
circuit.cx(1,2)

# control X-gate Q0 -> Q1 and hadamard gate on Q0
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()

# measurements on Q0 and Q1
circuit.measure([0,1], [0,1])
circuit.barrier()

circuit.cx(1,2)
circuit.cz(0,2)
circuit.measure(2,2)
circuit.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = simulator.run(circuit, shots=1000).result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```
